```python
def count_ordinal(df, feature):
    count = df.value_counts(feature)
    count.plot(kind='bar')
```

```python
count_ordinal(df, 'bg')
```

```python
count_ordinal_sns(df, 'bg')
```

```python
def outliers_to_nan(df, features=[], n=3):
    df_to_clean = df.copy()
    # print(df_to_clean)
    # print(df_to_clean.std())
    filtered_df = df_to_clean.mask(abs(df_to_clean - df_to_clean.mean()) > n * df_to_clean.std(), np.nan)
    return filtered_df
```

```python
df_no_outliers = outliers_to_nan(df_nan)

df_no_outliers
```
