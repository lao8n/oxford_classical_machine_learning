```python
def outliers_to_nan(df, features=[], n=3):
    df_to_clean = df.copy()
    # print(df_to_clean)
    # print(df_to_clean.std())
    filtered_df = df_to_clean.mask(abs(df_to_clean - df_to_clean.mean()) > n * df_to_clean.std(), np.nan)
    return filtered_df
```

```python
def impute_nan_values(df, feature, imputation_type='mean'):
    df_imputed = df.copy()
    if imputation_type == 'mean':
        imputation_value = df_imputed[feature].mean()
    elif imputation_type == 'median':
        imputation_value = df_imputed[feature].median()
    else: 
        raise ValueError('Invalid imputation type. Use mean or median')
    df_imputed[feature].fillna(imputation_value, inplace=True)
    return df_imputed
```

```python
df_imputed = impute_nan_values(df_nan, 'T1', 'mean')
df_imputed = impute_nan_values(df_imputed, 'T2', 'mean')
df_imputed = impute_nan_values(df_imputed, 'T3', 'mean')
df_imputed = impute_nan_values(df_imputed, 'T4', 'mean')
mn.matrix(df_imputed)
```

```python
for v in ['m1_cost', 'm2_cost', 'm3_cost', 'm4_cost']:
    df_imputed = impute_nan_values(df_imputed, v, 'mean')
for v in ['V1','V2', 'V3', 'V4', 'A1', 'A2', 'A3', 'A4', 'B1', 'B2', 'B3', 'B4', 'P1', 'P2', 'P3', 'P4']:
    df_imputed = impute_nan_values(df_imputed, v, 'median')
mn.matrix(df_imputed)
```

```python
num_columns = df.columns[2:-1]
df_imputed = pd.DataFrame(imputed, columns=num_columns)
df_imputed['ag'] = df_no_outliers['ag']
df_imputed['bg'] = df_no_outliers['bg']
mn.matrix(df_imputed)
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy='most_frequent')
df_imputed[['ag', 'bg']] = imp.fit_transform(df_imputed[['ag', 'bg']])
mn.matrix(df_imputed)
```

```python
features_to_be_standardized = num_columns[:-8]
plot_box(df_imputed, features_to_be_standardized)
```

```python
from sklearn.preprocessing import StandardScaler
def feature_standardize(df, features_to_standardize):
    scaler = StandardScaler()
    array = scaler.fit_transform(df[features_to_standardize])
    return pd.DataFrame(array, columns=features_to_standardize), scaler.mean_, scaler.scale_
df_scaled, scaler_mean, scaler_scale = feature_standardize(df_imputed, features_to_be_standardized)
plot_box(df_scaled, features_to_be_standardized)
print(scaler_mean)
print(scaler_scale)
```

```python
df_scaled[num_columns[-8:]] = df_no_outliers[num_columns[-8:]]
df_scaled[['ag', 'bg']] = df_no_outliers[['ag', 'bg']]
```

```python
ag_mapping = {'High': 3, 'Medium': 2, 'Low': 1}
def transform_ordinal_to_numeric(df, feature, mapping_dict):
    df_transformed = df.copy()
    df_transformed[feature] = df_transformed[feature].map(mapping_dict)
    return df_transformed
df_transformed = transform_ordinal_to_numeric(df_scaled, 'ag', ag_mapping)
print(df_transformed['ag'])
```

```python
df_final = pd.get_dummies(df_transformed, columns=['bg'])
df_final.loc[:, 'bg_Argon':'bg_Neon']
```

```python
import matplotlib.pyplot as plt
for column in df_final.columns:
    plt.figure(figsize=(12, 2))
    sns.kdeplot(df_final[column], label=column)
    plt.title(f'KDE Plot for {column}')
    plt.legend()
```

```python
features_to_plot = ['m1_cost', 'm2_cost', 'm3_cost', 'm4_cost']
fig, axes = plt.subplots(nrows=len(features_to_plot),
                         figsize=(12, 2 * len(features_to_plot)), 
                         sharex=True)
for i, column in enumerate(features_to_plot):
    sns.kdeplot(df_final[column], 
                label=column,
                ax=axes[i],
                shade=True,
                color="orange")
axes[i].set_title(f'KDE Plot for {column}')
axes[i].legend()
```
